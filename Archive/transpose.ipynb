```python
import numpy as np
import pandas as pd
```

```python
mut_pd = pd.read_csv('../API/API_datasets/ge_numerical.csv', sep =",")
mut_header_pd = pd.read_csv('../datasets_498/geneexp_entrezid.csv', sep =",", names=["entrez"])
```

```python
entrez_pd = pd.read_csv('../../EntrezID.txt', skiprows = [0], sep ="\t", names = ["Hugo", "Symbol", "Entrez_ID"]) 
entrez_pd.head()
```

```python
mut_header_pd.head()
```

```python
hugo_lst = []
for symbol in mut_header_pd["entrez"]:
    entrez_index = entrez_pd.index[entrez_pd['Entrez_ID'] == symbol].tolist()
    if entrez_index != []:
        entrez_ID = entrez_pd.get_value(entrez_index[0], "Symbol")
        hugo_lst.append(entrez_ID)
    else:
        hugo_lst.append([])
mut_header_pd["Entrez_ID"] = hugo_lst
mut_header_pd = mut_header_pd.drop(["entrez"], axis=1)
mut_header_pd.head()
```

```python
mut_pd["hugo"] = mut_header_pd
```

```python
mut_pd = mut_pd.transpose()
```
