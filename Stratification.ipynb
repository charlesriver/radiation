```python
for index, row in hist_pd.iterrows():
    if hist_pd.loc[index,"Sites"] == "lung":
        if hist_pd.loc[index,"Subhistology"] == "bronchioloalveolar_adenocarcinoma" or hist_pd.loc[index,"Subhistology"] == "mixed_adenosquamous_carcinoma":
            hist_pd.loc[index,"Sites"] = "adenocarcinoma"
        elif hist_pd.loc[index,"Subhistology"] == "mucoepidermoid_carcinoma":
            hist_pd.loc[index,"Sites"] = "squamous_cell_carcinoma"
        else:
            hist_pd.loc[index,"Sites"] = hist_pd.loc[index,"Subhistology"]
hist_pd.head()
```

```python
# encode categorical variables
def encode_categorical(array):
    if not array.dtype == np.dtype('float64'):
        return preprocessing.LabelEncoder().fit_transform(array) 
    else:
        return array

# apply one hot encoding
def one_hot (dataframe): 
    sites_lst = []
    for index, val in enumerate(hist_pd["Cclid"]):
        if val in list(dataframe.index):
            sites_lst.append(hist_pd.loc[index,"Sites"])
    dataframe.loc[:,"Sites"] = sites_lst
    
    # Categorical columns for use in one-hot encoder
    categorical = (dataframe.dtypes.values != np.dtype('float64'))

    # Encode all labels
    dataframe = dataframe.apply(encode_categorical)
    dataframe_np = dataframe.as_matrix()
    return dataframe_np, dataframe
```
