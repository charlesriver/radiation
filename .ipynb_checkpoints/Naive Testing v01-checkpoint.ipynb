```python
import numpy as np
import pandas as pd
import scipy as sp
import pickle
from scipy import stats 
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVR
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_name = "datamaster"
file_object = open(file_name,'r')  
compiled_pd = pickle.load(file_object)
compiled_pd.tail(n=10)
```

```python
no_auc_pd = compiled_pd.drop(compiled_pd.index[len(compiled_pd)-1])
no_auc_trans_pd = no_auc_pd.transpose()
no_auc_trans = no_auc_trans_pd.as_matrix()
print np.shape(no_auc_trans)
no_auc_trans_pd.tail(n=10)
```

```python
auc_pd = compiled_pd.iloc[-1:]
auc = auc_pd.as_matrix()
auc = auc.reshape([472,1])
print np.shape(auc)
auc_pd
```

```python
linreg = LinearRegression()
svr = SVR()
rf = RandomForestRegressor()

# test naive models
def test_model (model):
    clf = model
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    r2 = score(clf, x_test, y_test)
    return y_pred, r2
```

```python
x_train, x_test, y_train, y_test = train_test_split(no_auc_trans, auc, test_size=0.5, random_state=42)
y_test = y_test.flatten()
y_train = y_train.flatten()

# calculated r_2 values
linreg_y_pred, linreg_r2 = test_model (linreg)
svr_y_pred, svr_r2 = test_model (svr)
rf_y_pred, rf_r2 = test_model (rf)
```

```python
#Function for computing the accuracy a given model on the entire test set
score = lambda model, x_test, y_test: pd.Series([model.score(x_test, y_test)],
                                                index=['Naive', 'Somewhat informed', 'Informed'])
score_df = pd.DataFrame({'Linear Regression': linreg_r2, 
                         'Support Vector Machine': svr_r2,
                         'Random Forest': rf_r2})
score_df
```

```python
def linreg (x_val, y_val):
    slope, intercept, r_val, p_val, stddev = stats.linregress(x_val, y_val)
    lin_space = np.linspace(np.min(x_val), np.max(x_val), 500)
    reg_y = [slope * val + intercept for val in lin_space]
    linreg_plt = plt.plot(lin_space, reg_y, color = "black")
    return linreg_plt

def draw_plot (y_pred, title, y_score):
    y_pred = y_pred.flatten()
    linreg_plt = linreg(y_pred, y_test)
    plt.scatter(y_pred, y_test)
    plt.title("%s r2 accuracy: %s" %(title, round(y_score[0], 3)))
    plt.xlabel("Predicted AUC")
    plt.ylabel("Actual AUC")
    plt.show()
```

```python
draw_plot (linreg_y_pred, "Linear Regression", linreg_r2)
draw_plot (svr_y_pred, "SVM", svr_r2)
draw_plot (rf_y_pred, "Random Forest", rf_r2)
```
