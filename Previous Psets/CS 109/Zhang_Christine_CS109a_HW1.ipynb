```python
# create a random list of numbers of sorting children data
from random import shuffle
random_children = np.random.permutation(len(df_children))

# split the children dataset into testing and training
training = pd.DataFrame(columns=categorical_variables)
testing = pd.DataFrame(columns=categorical_variables)
for i in range(0, len(df_children)):
    if random_children[i] < 240:
        testing.loc[i] = (df_children.iloc[i,:])
    else:
        training.loc[i] = (df_children.iloc[i,:])

# create separate arrays storing marker_1 and marker_2 values for children testing data
xpoint = []
ypoint = []
for n in range(0, len(testing)):
    xpoint.append(testing.iloc[n, 0])
    ypoint.append(testing.iloc[n, 1])

# classify by comparing the testing data against every data point in the training dataset
def classify(training, testing):
    all_marker_1 = []
    all_marker_2 = []
    patient = []
    for entry in range(0, len(training)):
        all_marker_1.append(training.iloc[entry, 0])
        all_marker_2.append(training.iloc[entry, 1])
        
    # Find the shortest Euclidean distance and assign the testing data point the subtype of the nearest data point
    for entry in range(0, len(xpoint)):
        temporary = []
        for each in range(0, len(all_marker_1)):
            temporary.append(((xpoint[entry] - all_marker_1[each]) ** 2 + (ypoint[entry] - all_marker_2[each]) ** 2) ** 0.5)
        minimum = min(temporary)
        index = temporary.index(minimum)
        index_value = training.iloc[index, :].values[2]
        patient.append(index_value)
    return patient

# Capture the new subtypes of the testing data
values = classify(training, testing)

# evaluate the accuracy of the classify function by comparing the assigned subtypes to the actual subtypes
def evaluate(testing):
    count = 0
    for n in range(0, len(testing)):
        if (testing.iloc[n, 2]) == values[n]:
            count = count + 1
        else:
            count = count
    print count / float(len(testing)) * 100, "%"

evaluate(testing)
```
