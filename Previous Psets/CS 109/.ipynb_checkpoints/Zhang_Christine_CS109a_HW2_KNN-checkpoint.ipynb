**Solution:**

```python
def split(data, m):
    
    # Create a random shuffled list of integers
    from random import shuffle
    random_children = np.random.permutation(len(data))
    
    # Use the random list to sort the children data into testing and training subsets
    test, train = [], []
    for i in range(0, len(data)):
        if random_children[i] < m * len(data):
            train.append(data[i])
        else:
            test.append(data[i])
    
    # Return the two split datasets
    return test, train
```

```python
def knn_predict(k, train, test):
    output = []
    
    # Iterate through list and append the absolute value 
    for i in range(len(test)):
        x_values_predict = []
        for each in range(len(train)):
            x_values_predict.append([abs(train[each][0] - test[i][0]), train[each][1]])
            
            # Sort from smallest to largest and take the first k elements
            x_values_predict.sort()
            k_elements = x_values_predict[:k]
            
            # Find the corresponding y values and average to get the estimated predicted value
            y_values = [b for [a,b] in k_elements]
            predicted_values = np.mean(y_values)
        output.append([test[i][0], predicted_values])
    return output
```

```python
def values(train):
    product, x_values, y_values, x_squared, y_squared = [], [], [], [], []
    
    # Separate the x values and y values and take the average
    for each in range(len(train)):
        x_values.append(train[each][0])
        y_values.append(train[each][1])
    x_average = np.mean(x_values)
    y_average = np.mean(y_values)
    
    # Calculate the sum of the products of x - x_average and y - y_average and the sum of x values squared
    for each in range(len(train)):
        product.append([(train[each][0] - x_average) * (train[each][1] - y_average)])
        x_squared.append([(train[each][0] - x_average)**2])
    product_total = (np.sum(product, axis = 0))[0]
    x_squared_sum = (np.sum(x_squared, axis = 0))[0]
    length = len(train)
    return float(product_total), float(x_squared_sum), float(x_average), float(y_average), float(length)
    
# Calcuate the linear regression slope and y-intercept
def linear_reg_fit(train):
    product_total, x_squared_sum, x_average, y_average, length = values(train)
    m = product_total / x_squared_sum
    b = y_average - m * x_average
    return m, b
```

```python
def linear_reg_predict(test, m, b):
    output = []
    for each in range(len(test)):
        
        # Use m and b from linear_reg_fit to predict y values given x 
        output.append([test[each][0], m * test[each][0] + b])
    return output
```

```python
def score(actual, predicted):
    
    # Find average y value
    y_values = [c for [a,c] in actual]
    y_average = np.mean(y_values)
    
    # Calcuate SSR and TSS for dataset
    SSR, TSS = [], []
    for each in range(len(predicted)):
        SSR.append((actual[each][1] - predicted[each][1])**2)
        TSS.append((actual[each][1] - y_average)**2)
    SSR_sum = np.sum(SSR)
    TSS_sum = np.sum(TSS)
    
    # Calculate R_squared given SSR and TSS values
    r_squared = 1 - (SSR_sum/TSS_sum)
    print r_squared
```

In this problem, we will be handling the problem of datasets with missing values. Clearly, we cannot simply remove entire rows or columns that contain missing values. In this problem, we explore two different ways to fill in missing values.

- Describe in detail how predictive models for data (like KNN and simple linear regression) can be used to fill in missing values in a data set.


- Implement your scheme. That is, write code (preferably a function ``fill`` or two functions ``fill_knn``, ``fill_lin_reg``), which takes an ``n x 2`` dataframe or array with values missing in the 2nd column and fills in these values using KNN and linear regression.

**Solution:**

```python
#--------  fill_knn
#input: missing_df (nx2 dataframe, some rows have missing y-values), 
#       full_df (nx2 dataframe, all x-values have correct y-values), 
#       no_y_ind (indices of missing values in missing_df), 
#       with_y_ind (indices of non-missing values in missing_df), 
#       k (integer)
#output: predicted_df (nx2 dataframe, first column is x-vals from missing_df, second column is predicted y-vals), 
#        r (float)

def fill_knn(missing_df, full_df, no_y_ind, with_y_ind, k):
    #preparing data in array form
    
    #training data
    x_train = missing_df.loc[with_y_ind, 'x'].values
    x_train = x_train.reshape((len(with_y_ind), 1)) #make x_train array into a 2D array
    y_train = missing_df.loc[with_y_ind, 'y'].values
    
    #testing data
    x_test = missing_df.loc[no_y_ind, 'x']
    x_test = x_test.values.reshape((len(no_y_ind), 1)) #make x_test array into a 2D array
    y_test = full_df.loc[no_y_ind, 'y'].values
    
    #fit knn model
    neighbours = KNN(n_neighbors=k)
    neighbours.fit(x_train, y_train)
    
    #predict y-values
    predicted_y = neighbours.predict(x_test)
    
    #score predictions
    r = neighbours.score(x_test, y_test)
    
    #fill in missing y-values
    predicted_df = missing_df.copy()
    predicted_df.loc[no_y_ind, 'y'] = pd.Series(predicted_y, index=no_y_ind)
    
    return predicted_df, r


#--------  fill_ling_reg
#input: missing_df (nx2 dataframe, some rows have missing y-values), 
#       full_df (nx2 dataframe, all x-values have correct y-values), 
#       no_y_ind (indices of missing values in missing_df), 
#       with_y_ind (indices of non-missing values in missing_df), 
#       k (integer)
#output: predicted_df (nx2 dataframe, first column is x-vals from missing_df, second column is predicted y-vals), 
#        r (float)


def fill_lin_reg(missing_df, full_df, no_y_ind, with_y_ind):
    #preparing data in array form
    
    #training data
    x_train = missing_df.loc[with_y_ind, 'x'].values
    x_train = x_train.reshape((len(with_y_ind), 1)) #make x_train array into a 2D array
    y_train = missing_df.loc[with_y_ind, 'y'].values
    
    #testing data
    x_test = missing_df.loc[no_y_ind, 'x'].values
    x_test = x_test.reshape((len(no_y_ind), 1)) #make x_test array into a 2D array
    y_test = full_df.loc[no_y_ind, 'y'].values
    
    #fit linear model
    regression = Lin_Reg()
    regression.fit(x_train, y_train)
    
    #predict y-values
    predicted_y = regression.predict(x_test)
    
    #score predictions
    r = regression.score(x_test, y_test)
    
    #fill in missing y-values
    predicted_df = missing_df.copy()
    predicted_df.loc[no_y_ind, 'y'] = pd.Series(predicted_y, index=no_y_ind)
    
    return predicted_df, r
```

KNN can be used to predict values by averaging the y values of the k nearest x neighbors. First you find the k data points closest to you and then you average their y values to get an estimation of the missing y values. This works on more uniform sets or sets with repeating patterns because it captures trends in data. 

Linear regression can be used to predict values by fitting the datapoints to a slope plus a y intercept. This works better on datasets with linear trends and is significantly less successful/accurate on quadratics, cubics, etc. We can plug in the missing indexes into the linear regression equation and use the outputed y value as our prediction.

## Challenge Problem: Combining Random Variables

This problem, we explore the distirbution of random variables that result from combining other random variables.

### Part (a): Adding Two Uniformly Distributed Variables

Consider the independent random variables $X\sim U(0, 1)$ and $Y\sim U(0, 1)$. Let $Z$ be the random variable $Z = X + Y$. 

What is the distribution of $Z$ (give the pdf for Z)? You should fully explain and support your conlusion. 

**Hint:** your solution can be a combination of experimentation, empirical evidence and/or algebra

**Solution:**

```python
# Create an array to occupy with output from simulation
z_array = []
def simulation():
    
    # Determine two random uniform variables
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    
    # Calculate z by summing up random variables
    z = x + y
    z_array.append(z)
    
# Run the simulation many times
for i in range(5000):
    simulation()

plt.hist(z_array, 40, color = "green")
plt.show()
```

Z ranges in values between [0, 2]. Z is equal to 

$$\int_R \! f(x) * f(q - x) \, \mathrm{d}x$$


where x is between [0, 2]. For x between [0,1], the integral reduces to 


$$\int_0^x \! 1 \, \mathrm{d}x = x$$

This gives us x for 0 <= x <= 1. For x between [1,2], the integral becomes 

$$\int_q^1 \! 1 \, \mathrm{d}x = 2 - x$$

$$where \ q = 1-x$$

This simplifies to 2-x for x between [1,2]. 


The distribution of Z therefore is:
$$x \ , \ 0 \leq x \leq 1$$
$$2-x \ , \ 1 \leq x \leq 2$$

### Part (b): Adding Multiple Uniformly Distributed Variables

Consider three independent random variables $X_1, X_2, X_3 \sim U(0, 1)$. Let $Z$ be the random variable $Z = X_1 + X_2 + X_3$. 

What is the distribution of $Z$? What if you add 10 or 12 independent (standard) uniformly distributed variables? Conjecture on the distribution of 
$$
Z = \lim_{n\to \infty} \sum_{i=1}^n X_i
$$
where $\left\{X_i \right\}$ are independent (standard) uniformly distributed variables.

**Hint:** your solution can be a combination of experimentation, empirical evidence and/or algebra

**Solution:**

```python
# Create an array to occupy with output from simulation
z_array = []
def simulation():
    
    # Create three random uniform variables
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    n = np.random.uniform(0,1)
    
    # Determine z by summing up all the random variables
    z = x + y + n
    z_array.append(z)

# Run the simulation many times
for i in range(5000):
    simulation()

plt.hist(z_array, 40, color = "green")
plt.show()
```

The pdf of Z with three random uniform variables is:

$$1/2x^2 \ , \ 0\leq x \leq1$$
$$1/2(-2x^2 + 6x - 3) \ , \ 1 \leq x \leq 2$$
$$1/2(x^2 -6x + 9) \ , \ 2 \leq x \leq 3$$

This can be derived from the probability density function of the distribution given by: 

$$f(x, n) =  \frac{1}{2(n-1)!} \sum_{k=0}^{n} (-1)^{k} {{N}\choose{k}} (x-k)^{n-1}sgn(x-k) $$

Where:

$$
 sgn(x-k) =
  \begin{cases} 
      \hfill -1  & x < k \\
      \hfill 0 & x = k \\
      \hfill 1 & x > k \\
  \end{cases}
$$


Source: http://www.math.uah.edu/stat/special/IrwinHall.html

```python
# Create an array to occupy with output from simulation
z_array = []
def simulation():
    
    # Create 12 random uniform variables
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    n = np.random.uniform(0,1)
    n1 = np.random.uniform(0,1)
    n2 = np.random.uniform(0,1)
    n3 = np.random.uniform(0,1)
    n4 = np.random.uniform(0,1)
    n5 = np.random.uniform(0,1)
    n6 = np.random.uniform(0,1)
    n7 = np.random.uniform(0,1)
    n8 = np.random.uniform(0,1)
    
    # Calculate z as the sum of all random variables
    z = x + y + n + n1 + n2 + n3 + n4 + n5 + n6 + n7 + n8
    z_array.append(z)

# Run the simulation many times
for i in range(10000):
    simulation()

plt.hist(z_array, 40, color = "green")
plt.show()
```

The distribution becomes more and more normally distributed with additional uniform random variables added. The distribution is centered at half the number of variables added. 

### Part (c): Combining Normally Distributed Variables

Consider the independent random variables $X\sim \mathcal{N}(0, 1)$ and $Y\sim \mathcal{N}(0, 1)$. Let $Z$ be the random variable $Z = X + Y$. 

What is the distribution of $Z$ (give the pdf for Z)? You should fully explain and support your conlusion. 

**Hint:** use properties of expected value and some experimentation.

**Solution:**

```python
# Create an array to occupy with output from simulation
z_array = []
def simulation():
    
    # Create two random normal variables 
    x = np.random.normal(0,1)
    y = np.random.normal(0,1)
    
    # Calculate z as the sum of the random variables
    z = x + y
    z_array.append(z)

# Run the simulation many times
for i in range(10000):
    simulation()

plt.hist(z_array, 40, color = "green")
plt.show()
```

$$ X \sim N(\mu_x, \sigma_x^2) $$
$$ Y \sim N(\mu_y, \sigma_y^2) $$

We can derive the distribution of Z using characteristic functions. The sum of X + Y or Z can be determined as the product of two characteristic functions. 

For normal distributions: 
$$\varphi(t) = exp(it\mu - \frac{\sigma^2t^2}{2})$$

Thus
$$\varphi_{X+Y}(t) = \varphi_{X}(t)\varphi_{Y}(t) = exp(it\mu_X - \frac{\sigma_X^2t^2}{2})exp(it\mu_Y - \frac{\sigma_Y^2t^2}{2})$$
$$\varphi_{X+Y}(t) = exp(it(\mu_X + \mu_Y) - \frac{(\sigma_X^2 + \sigma_X^2)t^2}{2})$$

From this derviation, we can conclude: 

$$ Z \sim N(\mu_x + \mu_y, \sigma_x^2 + \sigma_y^2) $$

### Part (d): Product of Normally Distributed Variables

Is the product of two normally distributed variables a normally distributed variable? You should fully explain and support your conlusion. 

**Solution:**

```python
# Create an array to occupy with output from simulation
z_array = []
def simulation():
    
    # Create two random normal variables
    x = np.random.normal(0,1)
    y = np.random.normal(0,1)
    
    # Calculate z as the product of two random variables
    z = x * y
    z_array.append(z)

# Run the simulation many times
for i in range(10000):
    simulation()

plt.hist(z_array, 50, color = "green")
plt.show()
```

No, this distribution is not normal. We can see from the simulation a sharp peak at 0. This corresponds to the presence of the delta function in calcuating the distribution of the product of two normal random varibles. With additional runs, the peak becomes more and more defined and the distribution deviates from the classical bell shape of a Normal distribution. 

The distribution can be represented by: 
$$P_{XY}(u) = \int_{0}^{\infty}\int_{0}^{\infty} (\frac{\frac{e^{-x^2}}{2\sigma_x^2}}{\sigma_x\sqrt(2\pi)})(\frac{\frac{e^{-y^2}}{2\sigma_y^2}}{\sigma_y\sqrt(2\pi)})\delta(xy-u) dxdy$$

The delta function is a special function that is zero everywhere except at 0, where it has infinite value. The delta function integrates over all real numbers to 1. 

$$
\delta(x) =
  \begin{cases} 
      \hfill 0  & x \neq 0 \\
      \hfill \infty & x = 0 \\
  \end{cases}
$$


$$\int_R \! \delta(x) \, \mathrm{d}x = 1$$
