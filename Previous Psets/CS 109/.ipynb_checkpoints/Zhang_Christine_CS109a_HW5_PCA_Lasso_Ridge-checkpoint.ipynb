### Part 1(b). Build a classifier

So far, we have only learned models that distinguishes between two classes. Develop and implement a **simple and naive** method of distinguishing between the three digits in our reduced dataset using binary classifiers. 

```python
###Build a classifier to distinguish between 0 and 1

#Remove all instances of class 3
x_binary = x_reduced[y != 3, :]

#Remove all instances of class 3
y_binary = y[y != 3]

#Fit logistic regression model for 0 vs 1
logistic_01 = LogReg()
logistic_01.fit(x_binary, y_binary)

###Build a classifier to distinguish between 1 and 3

#Remove all instances of class 0
x_binary = x_reduced[y != 0, :]

#Remove all instances of class 0
y_binary = y[y != 0]

#Fit logistic regression model for 1 vs 3
logistic_13 = LogReg()
logistic_13.fit(x_binary, y_binary)

###Build a classifier to distinguish between 0 and 3

#Remove all instances of class 1
x_binary = x_reduced[y != 1, :]

#Remove all instances of class 1
y_binary = y[y != 1]

#Fit logistic regression model for 0 vs 3
logistic_03 = LogReg()
logistic_03.fit(x_binary, y_binary)

#Predict a label for our dataset using each binary classifier
y_pred_01 = logistic_01.predict(x_reduced)
y_pred_13 = logistic_13.predict(x_reduced)
y_pred_03 = logistic_03.predict(x_reduced)

#Now, for each image, we have THREE predictions!
#To make a final decision for each image, we just take a majority vote.
n = x_reduced.shape[0]
y_votes = np.zeros((n, 3))

#Votes for 0
y_votes[y_pred_01 == 0, 0] += 1
y_votes[y_pred_03 == 0, 0] += 1

#Votes for 1
y_votes[y_pred_01 == 1, 1] += 1
y_votes[y_pred_13 == 1, 1] += 1

#Votes for 3
y_votes[y_pred_03 == 3, 2] += 1
y_votes[y_pred_13 == 3, 2] += 1

#For each image, label it with the class that get the most votes
y_pred = y_votes.argmax(axis = 1)

#Relabel class 2 as class 3
y_pred[y_pred == 2] = 3

#Accuracy of our predictions
print 'Accuracy of combined model:', np.mean(y == y_pred)
```

```python
#--------  fit_and_plot_model
# A function to fit a binary LogReg model and visualize it
# Input: 
#      model (LogReg model)
#      ax (axes object for plotting)
#      legend_label (legend label for the plot)

def plot_model(model, ax, legend_label, color):
    #Get the coefficients from logistic regression model
    coef = model.coef_[0]
    intercept = model.intercept_
    
    #Find the max and min horizontal values of our data
    x_0 = np.min(x_reduced[:, 0])
    x_1 = np.max(x_reduced[:, 0])
        
    #Plug int the max and min horizontal values of our data into the equation
    #of the line defined by the coefficients
    y_0 = (-intercept - coef[0] * x_0) / coef[1]
    y_1 = (-intercept - coef[0] * x_1) / coef[1]

    #Plot a line through the pair of points we found above
    ax.plot([x_0, x_1], [y_0, y_1], label=legend_label, color=color)


fig, ax = plt.subplots(1, 1, figsize=(10, 6))

#Scatter plot of our data
ax.scatter(x_reduced[y==0, 0], x_reduced[y==0, 1], color='b', label='0')
ax.scatter(x_reduced[y==1, 0], x_reduced[y==1, 1], color='r', label='1')
ax.scatter(x_reduced[y==3, 0], x_reduced[y==3, 1], color='g', label='3')

#Plot decision boundaries for 0 vs 1
plot_model(logistic_01, ax, '0 vs 1', 'magenta')
#Plot decision boundaries for 1 vs 3
plot_model(logistic_13, ax, '1 vs 3', 'cyan')
#Plot decision boundaries for 0 vs 3
plot_model(logistic_03, ax, '0 vs 3', 'orange')

#Label graph axes and set limits to focus on data
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_xlim([np.min(x_reduced[:,0]), np.max(x_reduced[:,0])])
ax.set_ylim([np.min(x_reduced[:,1]), np.max(x_reduced[:,1])])
ax.set_title('our reduced data with all three decision boundaries (from binary classifiers)')
ax.legend()
plt.show()
```

```python
#Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = sk_split(x, y, test_size = 0.3)

#Calculate a polynomial basis for plotting
def polynomial_basis (x, degree):
    p = np.arange (1, degree + 1)
    return x[:, np.newaxis] ** p

# Create logistic regression object
logitm1 = LogReg(C = 1000000)
logitm1.fit (x, y)

# The coefficients
print('Estimated beta1: \n', logitm1.coef_)
print('Estimated beta0: \n', logitm1.intercept_)
```

```python
#Initialize arrays for testing the value of C
l1_accuracy, l2_accuracy, Cs, C_array = [], [], [], []

#Prepare C values to test
for i in range(-7, 8):
    C_array.append(10 ** i)
    
for i in C_array:
    C = i
    
    #Use KFold to split hte dataset into 5 folds
    kf = KFold(x_train.shape[0], n_folds = 5)
    
    #Implement l1 and l2 logistic regressions 
    reg_l1 = LogReg(C = C, penalty = "l1")
    reg_l2 = LogReg(C = C, penalty = "l2")
    
    #Calculate the average accuracy and store in arrays
    accuracy_l1 = np.mean(val_score(reg_l1, x_train, y_train, cv = kf))
    accuracy_l2 = np.mean(val_score(reg_l2, x_train, y_train, cv = kf))
    l1_accuracy.append(accuracy_l1)
    l2_accuracy.append(accuracy_l2)
    Cs.append(C)

#Plot the logistic regression accuracies on a semilog plot
plt.semilogx(10.0**np.arange(-7, 8), l1_accuracy,  color='red', label = "Lasso")
plt.semilogx(10.0**np.arange(-7, 8), l2_accuracy,  color='blue', label = "Ridge")
plt.xlabel ("C")
plt.ylabel("accuracy")
plt.legend(loc='best')
plt.show()
```

```python
#Find the max accuracy and index of the best value
max_accuracy = np.amax(l2_accuracy)
max_arg = np.argmax(l2_accuracy)
max_c = Cs[max_arg]
print 'Max accuracy: ', max_accuracy
print 'C Value of Max accuracy: ', max_c
```

```python
#Set C equal to the value that produces the maximum and perform logistic regression
C = max_c
logitm_l1 = LogReg(C = C, penalty = "l1")
logitm_l1.fit (x_train, y_train)
logitm_l2 = LogReg(C = C, penalty = "l2")
logitm_l2.fit (x_train, y_train)

#Print cofficients and impact of penalty
print(logitm_l1.coef_)
print 'Number of non-zero l1 predictors: ', len([i for i, item in enumerate(logitm_l1.coef_[0]) if abs(item) > 0])
print(logitm_l2.coef_)
print 'Number of non-zero l2 predictors: ', len([i for i, item in enumerate(logitm_l2.coef_[0]) if abs(item) > 0])
```

```python
#Read all the datasets
data_train = pd.read_csv('datasets/dataset_3_train.txt')
data_test_1 = pd.read_csv('datasets/dataset_3_test_1.txt')
data_test_2 = pd.read_csv('datasets/dataset_3_test_2.txt')

#Assign new columns
columns = ['Marker1', 'Marker2', 'Marker3', 'Marker4', 'Marker5', 'Results']
data_train.columns = columns
data_test_1.columns = columns
data_test_2.columns = columns

#Parse the predicts from the response variables for all datasets
x = data_train.iloc[:, :5]
y = data_train.iloc[:, -1]

x_test1 = data_test_1.iloc[:, :5]
y_test1 = data_test_1.iloc[:, -1]

x_test2 = data_test_2.iloc[:, :5]
y_test2 = data_test_2.iloc[:, -1]
```

```python
#Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = sk_split(x, y, test_size = 0.3)

#Create a polynomial basis for plotting
def polynomial_basis (x, degree):
    p = np.arange (1, degree + 1)
    return x[:, np.newaxis] ** p

# Create logistic regression object
logitm1 = LogReg(C = 1000000)
logitm1.fit (x, y)

#Print the resulting coefficients
print('Estimated beta1: \n', logitm1.coef_)
print('Estimated beta0: \n', logitm1.intercept_)
```

```python
l1_accuracy, l2_accuracy, Cs, C_array = [], [], [], []

#create C array of values
for i in range(-7, 8):
    C_array.append(10 ** i)

#perform kfold to test accuracy of logistic regression 
for i in C_array:
    C = i
    kf = KFold(x_train.shape[0], n_folds = 5)
    
    #test both l1 and l2 penalties on the regression
    reg_l1 = LogReg(C = C, penalty = "l1")
    reg_l2 = LogReg(C = C, penalty = "l2")
    accuracy_l1 = np.mean(val_score(reg_l1, x_train, y_train, cv = kf))
    accuracy_l2 = np.mean(val_score(reg_l2, x_train, y_train, cv = kf))
    l1_accuracy.append(accuracy_l1)
    l2_accuracy.append(accuracy_l2)
    Cs.append(C)

#plot the accuracy on a semilog scale
plt.semilogx(10.0**np.arange(-7, 8), l1_accuracy,  color='red', label = "Lasso")
plt.semilogx(10.0**np.arange(-7, 8), l2_accuracy,  color='blue', label = "Ridge")
plt.xlabel ("C")
plt.ylabel("accuracy")
plt.legend(loc='best')
plt.show()
```

```python
#find max accuracy value and index
max_accuracy = np.max(l2_accuracy)
max_arg = np.argmax(l2_accuracy)
max_c = Cs[max_arg]
print 'Max accuracy: ', max_accuracy
print 'C Value of Max accuracy: ', max_c
```

```python
C = max_c

#perform logistic regression with l1 and l2 penalties on the training datset
logitm_l1 = LogReg(C = C, penalty = "l1")
logitm_l1.fit (x_train, y_train)
logitm_l2 = LogReg(C = C, penalty = "l2")
logitm_l2.fit (x_train, y_train)

# print coefficients of both logistic models
print(logitm_l1.coef_)
print 'Number of non-zero l1 predictors: ', len([i for i, item in enumerate(logitm_l1.coef_[0]) if abs(item) > 0])
print(logitm_l2.coef_)
print 'Number of non-zero l2 predictors: ', len([i for i, item in enumerate(logitm_l2.coef_[0]) if abs(item) > 0])
```
