```python
def plot_lin_fit(m, b):
    # Read the dataset and plot pairs in a scatterplot
    dataset = pd.read_csv('data.csv')
    dataset_np = dataset.as_matrix()
    fig = plt.figure(figsize=(10, 10))
    dataset_plot = dataset.plot(kind='scatter', x='x', y='y', title ="Dataset 1 Full", color='Blue')

    # Use the given m and b values to produce a linear line of fit
    x = np.linspace(0, 10, 500)
    y = m * x + b;
    plt.plot(x, y, label = "m: " + str(m) + " b: " + str(b))

    plt.legend()
    plt.show()

    # Calcuate residual values of dataset from line of fit
    residual = []
    for each in range(len(dataset_np)):
        residual.append(dataset_np[each][1] - (m*dataset_np[each][0] + b))

    # Calculate total residual sum
    residual_sum = [a**2 for a in residual]
    total_sum = np.sum(residual_sum)

    # Calculate TSS and RSS for r squared calculation
    y_values = [[b] for [a,b] in dataset_np]
    y_average = np.mean(y_values)
    sum_of_squares = []
    for each in y_values:
        sum_of_squares.append((each - y_average)**2)
    total_sum_of_squares = np.sum(sum_of_squares)

    # Calculate r squared using TSS and RSS as determined above
    r_squared = 1 - total_sum/total_sum_of_squares
    print "R-squared: ", r_squared
```

```python
# Test the fit of line y = 30 - 2x
plot_lin_fit(-2, 30)
```

```python
def polynomial_regression_fit(x_train, y_train, degree):
    # Expand the array for x and populate columns with x raised to the power of that degree
    n = x_train.shape[0]
    x_expanded = np.zeros([n, degree])
    for d in range(1, degree + 1):
        x_expanded[:, d - 1] = np.power(x_train, d)
    
    # Fit sklearn multiple linear regression model
    reg = Lin_Reg()
    reg.fit(x_expanded, y_train)
    
    # Get coefficients from fitted model
    w = reg.coef_
    c = reg.intercept_
    
    return w, c

def polynomial_regression_predict(w, c, degree, x_test):
    # Expand the array for x and populate columns with x raised to the power of that degree
    n = x_test.shape[0]
    x_expanded = np.zeros([n, degree])
    for d in range(1, degree + 1):
        x_expanded[:, d - 1] = np.power(x_test, d)
        
    # Predict y-vals
    y_pred = np.dot(x_expanded, w) + c
        
    return y_pred


def polynomial_regression_score(y_pred, y_test):
    # calculate sq_error for r squared calculation
    sq_error = np.sum(np.square(y_test - y_pred))
    
    # Evaluate squared error for a predicting the mean value, against target labels
    y_mean = np.mean(y_test)
    y_variance = np.sum(np.square(y_test - y_mean))
    
    # Evaluate R^2 score value
    r_squared = 1 - sq_error / y_variance

    return r_squared, sq_error
```

```python
# use sk_split to divide up training and testing data
z_train, z_test, x_train, x_test = sk_split(z_values, x_values, test_size = 0.3)

def fit_degrees_and_plot(ax, x_train, y_train, x_test, y_test, n_degrees): 
    # Store train R^2 score values for different degrees
    r_squared_train = np.zeros(n_degrees) 
    
    # Store test R^2 score values for different degrees
    r_squared_test = np.zeros(n_degrees) 
    
    # Iterate over degrees
    for degree in range(1, n_degrees+1):
        # Fit polynomial model on training set
        w, c = polynomial_regression_fit(x_train, y_train, degree)
        
        # Evaluate model on training set
        y_pred = polynomial_regression_predict(w, c, degree, x_train)
        r_squared_train[degree - 1], _ = polynomial_regression_score(y_pred, y_train)
        
        # Evaluate model on test set
        y_pred = polynomial_regression_predict(w, c, degree, x_test)
        r_squared_test[degree - 1], _ = polynomial_regression_score(y_pred, y_test)
    
    # plot testing data and label graph
    ax.plot(np.arange(1,n_degrees + 1), r_squared_test, 'rx-', label='Test') 
    ax.set_xlabel('Degree of Polynomial')
    ax.set_ylabel('R^2 score')
    ax.legend(loc='lower right')
    
    # Calculate optimal degree by finidng maximium value
    best_deg = np.argmax(r_squared_test) + 1
    r_sq_max = max(r_squared_test)
    
    return ax, best_deg, r_sq_max

# Fit x values against polynomial regression and plot
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax, best_deg, r_sq_max = fit_degrees_and_plot(ax, z_train, x_train, z_test, x_test, 15)
plt.title('Model Fitness on X Data Testset')
plt.show()

# Print out the optimal polynomial degree and r squared value
print 'Best fit on X stest set at', best_deg
print 'Max R^2 score on test set', round(r_sq_max, 2)
```

```python
# Fit y values against polynomial regression and plot
z_train, z_test, y_train, y_test = sk_split(z_values, y_values, test_size = 0.3)
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax, best_deg, r_sq_max = fit_degrees_and_plot(ax, z_train, y_train, z_test, y_test, 15)
plt.title('Model Fitness on Y Data Testset')
plt.show()

# Print out the optimal polynomial degree and r squared value
print 'Best fit on Y test set at', best_deg
print 'Max R^2 score on test set', round(r_sq_max, 2)
```

```python
# Initalize constants for bootstrap, use 3 for degrees as calculated above
num_samples = 60
degree = 3
y_pred_array = []

# Perform bootstrap calculation to generate multiple y values
def boot_strap_r_squared(x_train, y_train, num_samples, degree):
    n = x_train.shape[0]
    
    for fold in range(0, num_samples):
        # Randomly a sample points from the train set of size 'n'
        indices = np.random.choice(np.arange(0, n), size=n)
        
        # obtain sample training sets with given indices
        x_train_sample = x_train[indices]
        y_train_sample = y_train[indices]
        
        # conduct polynomial regression fit and store y intercept
        w, c = polynomial_regression_fit(x_train_sample, y_train_sample, degree)
        y_pred_array.append(c)
    return y_pred_array, np.mean(y_pred_array), np.std(y_pred_array)
        
# Perform bootstrap on y values to obtain mean and standard deviation
y_pred_array, mean_y, std_y = boot_strap_r_squared(z_values, y_values, num_samples, degree)
```

```python
# Plot histogram of coefficients, and report their confidence intervals 
fig, axes = plt.subplots(figsize=(8, 5))

# Compute mean for the coefficent from subsamples
coef_mean = np.mean(y_pred_array)
print 'Mean Y Value: ', coef_mean
print "Standard Deviation of Y: ", std_y

# Compute confidence interval at 90% confidence level
conf_int_left = np.percentile(y_pred_array, 5.0)
conf_int_right = np.percentile(y_pred_array, 95.0)

# Plot histogram of coefficient values
axes.hist(y_pred_array, alpha=0.5)

# Plot vertical lines at mean and left, right extremes of confidence interval
axes.axvline(x = coef_mean, linewidth=3)
axes.axvline(x = conf_int_left, linewidth=1, c='r')
axes.axvline(x = conf_int_right, linewidth=1, c='r')

# Set plot labels
axes.set_title('90% Confidence Interval : [' + str(round(conf_int_left, 4)) 
                  + ', ' 
                  + str(round(conf_int_right, 4)) + ']')
axes.set_xlabel('Y Values')
axes.set_ylabel('Frequency')

plt.show()
```

The number of beds and bathrooms can be used to predict the number residents in the house. A house with 2 people needs a minimum of 2 beds. With more residents, there tends to be more bathrooms so they will not have to share the same ammenities. 

```python
# read population data and extract bed and bath predictors
popdata = pd.read_csv('pop_data.csv')
bed_bath = popdata.iloc[:, 1:3]

# split data based on filled in number of residents and NaN values
number = popdata['residents']
y_nonan = popdata[~np.isnan(number)]
y_nan = popdata[np.isnan(number)]

r_squared = []
for k in range(1,300):
    # use sk_split to separate completed data to use for training
    train, test = sk_split(y_nonan, test_size = 0.5)
    
    # separate data into training and testing sets for independent and dependent variables
    x_train = train.iloc[:,1:3]
    x_test = test.iloc[:,1:3]
    y_train = np.ravel(train.iloc[:,0:1])
    y_test = np.ravel(test.iloc[:,0:1])
    
    # perform KNN analysis using a variety of k values
    neighbours = KNN(n_neighbors = k)
    neighbours.fit(x_train, y_train)
    predicted_y = neighbours.predict(x_test)
    
    # Calculate r squared from analysis 
    r_KNN = neighbours.score(x_test, y_test)
    r_squared.append(r_KNN)

# Find optimal k value as maximum r squared index
best_deg = np.argmax(r_squared) + 1
r_sq_max = max(r_squared)

# Plot r squared for varying k values
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
plt.plot(range(1,300), r_squared)
plt.title('R squared values for varying k')
plt.show()

# Print out the optimal k and r squared values
print 'Best k on test set at', best_deg
print 'Max R^2 score on test set', round(r_sq_max, 2)
```

```python
# Obtain optimal k value from above analysis
k = best_deg

# Split completed data to perform a KNN fit on optimal k
train, test = sk_split(y_nonan, test_size = 0.5)
x_train = train.iloc[:,1:3]
x_test = y_nan.iloc[:,1:3]
y_train = np.ravel(train.iloc[:,0:1])
y_test = np.ravel(y_nan.iloc[:,0:1])
neighbours = KNN(n_neighbors = k)
neighbours.fit(x_train, y_train)

# Use original x_test data to predict the missing y values
predicted_y = neighbours.predict(x_test)
y_nan.iloc[:,0:1] = predicted_y

# Append newly predicted values to previous filled in values for whole dataset
full = y_nonan.append(y_nan)
```
